```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top == None

    def display(self):
        if self.is_empty():
            return "Stack is Empty"
        current = self.top
        while current:
            print(current.data)
            current = current.next
        print()

    def push(self,data):
        node = Node(data)
        node.next = self.top
        self.top = node
        self.display()

    def pop(self):
        if self.is_empty():
            return None
        popped = self.top.data
        temp = self.top
        self.top = temp.next
        temp.next = None
        self.display()
        print(f"Popped data is {popped} ")

    def peek(self):
        if self.is_empty():
            return None
        peeked = self.top.data
        self.display()
        print(f"Peeked data is {peeked} ")
```

```python
my_stack = Stack()
```

```python
my_stack.display()
```

```python
my_stack.is_empty()
```

```python
my_stack.push(10)
```

```python
my_stack.push(20)
my_stack.push(30)
my_stack.push(40)
my_stack.push(50)
```
