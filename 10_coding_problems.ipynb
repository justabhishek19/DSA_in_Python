# **Reverse A String Using A Stack**

```python
from collections import deque
```

```python
def reverse_str_using_stack(string):
    stack = deque()
    result = ""
    # push in stack
    for ch in string:
        stack.append(ch)

    # pop out from stack and store in result
    while stack:
        result += stack.pop()
    return result
```

```python
reverse_str_using_stack("abhishek")
```

```python
reverse_str_using_stack("god🤣")
```

# **Check Balanced Parenthesis**

```python
def is_balannced_parenthesis(expression):
    stack = deque()
    pairs = {
        ')':'(',
        ']':'[',
        '}':'{'
    }

    for ch in expression:
        if ch in '({[':
            stack.append(ch)
        elif ch in ')}]':
            if (not stack) or stack.pop() != pairs[ch]:
                return False

    return len(stack) == 0
```

```python
is_balannced_parenthesis('({[([()])]})')
```

```python
is_balannced_parenthesis('({[([()])]}])')
```

```python
is_balannced_parenthesis('[({[([()])]})')
```

# **Undo Operation Simulation**

```python
def simulate_undo_operation(commands):
    stack = deque()
    for command in commands:
        if command != 'undo':
            stack.append(command[-1])
        else:
            if stack:
                stack.pop()
            else:
                print("Cannot perform Undo Operation")
                return
    return stack
```

```python
commands = [
    'Page A',
    'Page B',
    'undo',
    'Page C',
    'undo',
    'undo',
    'undo'
]

simulate_undo_operation(commands)
```
