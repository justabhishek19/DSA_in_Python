```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class CSLL:
    def __init__(self):
        self.head = None
        self.length = 0

    def is_empty(self):
        return self.head is None

    def display(self):
        if self.is_empty():
            print("Empty List")

        else:
            current = self.head
            while True:
                print(current.data, end=" ")
                current = current.next
                if current == self.head:
                    break
            print()

    def insert_at_beginning(self,value):
        node = Node(value)
        if self.is_empty():
            self.head = node
            self.head.next = node
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            node.next = self.head
            current.next = node
            self.head = node
        self.length += 1
        self.display()

    def insert_at_end(self,value):
        node = Node(value)
        if self.is_empty():
            self.head = node
            self.head.next = node
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            node.next = self.head
            current.next = node
        self.length += 1
        self.display()

    def insert_at_index(self,pos,value):
        if pos == 0:
            self.insert_at_beginning(value)
            
        elif pos == self.length:
            self.insert_at_end(value)

        elif pos < 0 or pos > self.length:
            print("Invalid Index")
            
        else:
            node = Node(value)
            current = self.head
            for _ in range(pos-1):
                current = current.next
            node.next = current.next
            current.next = node
            self.display()
            self.length += 1

    def search(self,key):
        if self.is_empty():
            print("List is Empty")
        else:
            current = self.head
            index = 0
            while True:
                if current.data == key:
                    return index
                index += 1
                
                current = current.next
                if current == self.head:
                    return -1

    def delete_at_beginning(self):
        if self.is_empty():
            print("List is Empty")
            return
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = self.head.next
            self.head.next = None
            self.head = current.next
        self.display()
        self.length -= 1
        

    def delete_at_end(self):
        if self.is_empty():
            print("List is Empty")
            return
        elif self.length == 1:
            self.head = None
        else:
            current = self.head
            while current.next.next != self.head:
                current = current.next
            current.next.next = None
            current.next = self.head
        self.display()
        self.length -= 1

    def delete_at_indx(self,pos):
        if pos == 0:
            self.delete_at_beginning()
            
        elif pos == self.length-1:
            self.delete_at_end()

        elif pos < 0 or pos >= self.length:
            print("Invalid Index")
            
        else:
            current = self.head
            for _ in range(pos-1):
                current = current.next
            temp = current.next.next
            current.next.next = None
            current.next = temp
            self.display()
            self.length -= 1
```
