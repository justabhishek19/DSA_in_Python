```python
class MyStack:
    def __init__(self):
        self.stack = []

    def length(self):
        return len(self.stack)

    def is_empty(self):
        return self.length() == 0

    def display(self):
        if self.is_empty():
            print("Stack is Empty")
        for element in self.stack[::-1]:
            print(element)
        print()

    def push(self,data):
        self.stack.append(data)
        self.display()

    def pop(self):
        if self.is_empty():
            return None
            
        popped = self.stack.pop()
        self.display()
        print(f"Popped element: {popped} ")

    def peek(self):
        if self.is_empty():
            return None
        print(self.stack[-1])
```

```python
my_stack = MyStack()
```

```python
my_stack.is_empty()
```

```python
my_stack.pop()
```

```python
my_stack.peek()
```

```python
my_stack.display()
```
