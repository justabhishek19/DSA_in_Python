```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class SLL:
    def __init__(self):
        self.head = None

    def length(self):
        count = 0
        curr = self.head
        while curr:
            count += 1
            curr = curr.next
        return count

    def display(self):
        if not self.head:
            print("Empty List")
        else:
            curr = self.head
            while curr:
                print(curr.data, end = " ")
                curr = curr.next
            print()
    def insert_at_beginning(self,value):
        node = Node(value)
        if self.head:
            node.next = self.head
        self.head = node
        self.display()

    def insert_at_end(self, value):
        node = Node(value)

        curr = self.head
        if self.head:
            while curr.next:
                curr = curr.next
        curr.next = node
        self.display()

    def insert_at_indx(self,pos,value):
        node = Node(value)

        if pos == 0:
            self.insert_at_beginning(value)
        elif pos > self.length() or pos <0:
            print("Invalid Position")
        else:
            curr = self.head
            for i in range(pos-1):
                curr = curr.next
    
            node.next = curr.next
            curr.next = node
            self.display()

    def search(self,key):
        pos = 0
        current = self.head

        if self.head:
            while current:
                if current.data == key:
                    return pos
                pos += 1
                current = current.next
            return -1
        else:
            print("List is Empty")

    def delete_at_beginning(self):
        if self.head:
            current = self.head
            self.head = self.head.next
            current.next = None
            self.display()
        else:
            print("Can't delete,list is empty")

    def delete_at_end(self):
    
        if self.head:
            if self.length() == 1:
                self.head = None
            else:
                curr = self.head
                while curr.next.next:
                    curr = curr.next
                curr.next = None
            self.display()
        else:
            print("Can't delete,List is Empty")


    def delete_at_indx(self,pos):

        if pos == 0:
            self.delete_at_beginning()
            
        elif pos >= self.length() or pos <0:
            print("Invalid Position")
            
        else:
            p = None
            q = self.head
            for i in range(pos):
                p = q
                q = q.next
            p.next = q.next
            q.next = None
            
            self.display()
```
