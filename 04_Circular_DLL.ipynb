```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
class CDLL:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
    def is_empty(self):
        if not self.length:
            return True
        return False

    def display_forward(self):
        if self.is_empty():
            print("List is Empty")
            
        current = self.head
        while True:
            print(current.data, end =" " )
            current = current.next
            if current == self.head:
                break
        print()

    def display_backward(self):
        if self.is_empty():
            print("List is Empty")
            
        current = self.tail
        while True:
            print(current.data, end =" ")
            current = current.prev
            if current == self.tail:
                break
        print()

    def insert_at_beginning(self,value):
        node = Node(value)
        if  self.is_empty():
            self.head = self.tail = node
            self.head.prev = node
            self.tail.next = node
        else:
            self.tail.next = node
            node.prev = self.tail
            
            node.next = self.head
            self.head.prev = node
            self.head = node
            
        self.length += 1
        self.display_forward()

    def insert_at_end(self,value):
        node = Node(value)
        if  self.is_empty():
            self.head = self.tail = node
            self.head.prev = node
            self.tail.next = node
        else:
            self.head.prev = node
            node.next = self.head
            self.tail.next = node
            node.prev = self.tail
            self.tail = node
        self.length += 1
        self.display_forward()

    def insert_at_indx(self,pos,value):
        if pos < 0 or pos > self.length:
            print('Invalid Index')
        elif pos == 0:
            self.insert_at_beginning(value)
        elif pos == self.length:
            self.insert_at_end(value)
        else:
            node = Node(value)
            current = self.head
            for _ in range(pos-1):
                current = current.next

            node.next = current.next
            current.next.prev = node
            current.next = node
            node.prev = current
            self.length += 1
            self.display_forward()

    def search(self,key):
        if self.is_empty():
            return 'List is Empty'
        else:
            current = self.head
            count = 0
            while True:
                if current.data == key:
                    print(count)
                    return
                count += 1
                current = current.next
                if current == self.head:
                    break
            return -1

    def delete_at_beginning(self):
        if  self.is_empty():
            return "Can't Delete, List is Empty."
        elif self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.head.prev = None
            self.tail.next = None
            self.head = self.head.next
            self.head.prev.next = None
            self.head.prev = self.tail
            self.tail.next = self.head
        self.length -= 1
        self.display_forward()

    def delete_at_end(self):
        if  self.is_empty():
            return "Can't Delete, List is Empty."
        elif self.length == 1:
            self.head = None
            self.tail = None
        else:
            temp = self.tail.prev
            
            self.tail.next = None
            self.tail.prev = None
            self.tail = temp
            self.tail.next = self.head
            self.head.prev = self.tail
        self.length -= 1
        self.display_forward()

    def delete_at_indx(self,pos):
        if  self.is_empty():
            return "Can't Delete, List is Empty."
        elif pos < 0 or pos >= self.length:
            return 'Invalid Index Man'
        elif pos == self.length-1:
            self.delete_at_end()
        elif pos == 0:
            self.delete_at_beginning()
        else:
            if self.length == 1:
                self.head = None
                self.tail = None
            else:
                current = self.head
                for _ in range(pos-1):
                    current = current.next
    
                temp = current.next.next
                current.next.prev = None
                current.next.next = None
                current.next = temp
                temp.prev = current
            self.length -= 1
            self.display_forward()
        
```
