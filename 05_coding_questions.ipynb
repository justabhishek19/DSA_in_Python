# Find The Middle Element of a Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.length = 0

    def display(self):
        if not self.head:
            print("Empty List")
        else:
            curr = self.head
            while curr:
                print(curr.data, end = " ")
                curr = curr.next
            print()
            
    def insert_at_beginning(self,value):
        node = Node(value)
        if self.head:
            node.next = self.head
        self.head = node
        self.length += 1
        self.display()

    def middle_element(self):
        n = self.length
        mid = n//2

        current = self.head
        for _ in range(mid):
            current = current.next

        print(f'The middle element is: {current.data}')
            
    
```

```python
sll = SinglyLinkedList()
```

```python
sll.insert_at_beginning(10)
```

```python
sll.insert_at_beginning(20)
sll.insert_at_beginning(30)
sll.insert_at_beginning(40)
sll.insert_at_beginning(50)
sll.insert_at_beginning(60)
```

```python
sll.middle_element()
```

```python
sll.insert_at_beginning(70)
```

```python
sll.middle_element()
```

# Count the number of Nodes in a Circular Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class CircularLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        current = self.head
        while True:
            print(current.data, end=" ")
            current = current.next
            if current == self.head:
                break
        print()

    def insert_at_beginning(self,value):
        node = Node(value)
        if not self.head:
            self.head = node
            self.head.next = node
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            node.next = self.head
            current.next = node
            self.head = node
        self.display()

    

    def total_nodes(self):
        count = 0
        current = self.head

        if not current:
            return 0
        while True:
            count += 1
            current = current.next
            if current == self.head:
                break
        print(f'Total number of nodes in a Circular Linked List is: {count}')
    
```

```python
cll = CircularLinkedList()
```

```python
cll.total_nodes()
```

```python
cll.insert_at_beginning(10)
```

```python
cll.total_nodes()
```

```python
cll.insert_at_beginning(20)
cll.insert_at_beginning(30)
cll.insert_at_beginning(40)
cll.insert_at_beginning(50)
```

```python
cll.total_nodes()
```

# Find the sum of all Nodes in a Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        if not self.head:
            print("Empty List")
        else:
            curr = self.head
            while curr:
                print(curr.data, end = " ")
                curr = curr.next
            print()
            
    def insert_at_beginning(self,value):
        node = Node(value)
        if self.head:
            node.next = self.head
        self.head = node
        self.display()

    def sum_of_all_nodes(self):
        current = self.head
        sum = 0
        
        while current:
            sum += current.data
            current = current.next
        print(f'The sum of all nodes is: {sum}')
```

# Find maximum and minimum element in a list

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        if not self.head:
            print("Empty List")
        else:
            curr = self.head
            while curr:
                print(curr.data, end = " ")
                curr = curr.next
            print()
            
    def insert_at_beginning(self,value):
        node = Node(value)
        if self.head:
            node.next = self.head
        self.head = node
        self.display()

    def max_min(self):
        if not self.head:
            return 'List is empty'
            
        current = self.head
        maxx = minn = current.data
    
        while current:
            if current.data > maxx:
                maxx = current.data
            if current.data < minn:
                minn = current.data
            current = current.next
        print(f"The maximum Element in a Linked List is {maxx} and minimum element is {minn}.")
```

```python
sll.max_min()
```

```python
sll.max_min()
```
